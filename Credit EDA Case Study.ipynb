# Data Cleaning

### Treating missing value

```python
# finding % of missing values in all columns
(application_data.isna().mean().round(4) * 100).sort_values(ascending=False)
```

```python
# filter out the columns having more than 50% null values
application_data = application_data.loc[:, (application_data.isna().mean().round(4) * 100 < 50)]
(application_data.isna().mean().round(2) * 100).sort_values(ascending=False)
```

```python
# after removing columns with very high null values, 
# now there are few columns like (EXT_SOURCE_3,AMT_REQ_CREDIT_BUREAU_QRT,AMT_REQ_CREDIT_BUREAU_YEAR,AMT_REQ_CREDIT_BUREAU_HOUR,
# AMT_REQ_CREDIT_BUREAU_DAY,AMT_REQ_CREDIT_BUREAU_WEEK,AMT_REQ_CREDIT_BUREAU_MON ) which are having less percentage(14%-20%) 
# of null values, all these columns are quantitative and null values can be imputed with their respective mean.
```

```python
columns_low_null=['EXT_SOURCE_3','AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_HOUR',
'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON']

application_data[columns_low_null]=application_data[columns_low_null].fillna(application_data[columns_low_null].mean().round(0))
application_data[columns_low_null].head(10)
```

### Changing datatypes of required columns

```python
# change column SK_ID_CURR to string as it is an identification number
application_data['SK_ID_CURR'] = application_data['SK_ID_CURR'].astype(int).astype('str')
```

```python
# converting data types of columns 'AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_HOUR',
#'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON' to integer as they all are 
# the Number of enquiries to Credit Bureau which can be integer only.

application_data[['AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_HOUR',
'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON']]=application_data[['AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_HOUR',
'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON']].astype(int)

application_data[columns_low_null].head(10)
```

```python
# Detecting and treating outliers
# for this we are taking here 3 numeric columns CNT_CHILDREN, AMT_INCOME_TOTAL,AMT_CREDIT

sns.boxplot(x=application_data['AMT_INCOME_TOTAL'])
# here total income has outliers present
```

### Finding top 10 correlations

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop
def get_top_abs_correlations(df, n):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

```
